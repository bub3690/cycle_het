# BRIXIA

```python
# trainset 부터


brixia_all_data = pd.read_csv('/hdd/project/cylce_het/dataset/processed_df/metadata_global_v2_process.csv')
brixia_consensus_data = pd.read_csv('/hdd/project/cylce_het/dataset/processed_df/metadata_consensus_v1.csv')

# consensus_csv Filename과 겹치는 데이터 제거
label_data = brixia_all_data[~brixia_all_data['Filename'].isin(brixia_consensus_data['Filename'])]

```

```python
# metadata subject id 따기 위해서.
brixia_metadata = pd.read_csv('/hdd/project/cylce_het/dataset/processed_df/metadata_global_v2_all.csv')
```

```python

# brixia1,brixia2,3,4,5,6 -> str로 바꾸고 -> ,로 합치기
label_data['severity']=label_data['brixia1'].astype(str)+','+label_data['brixia2'].astype(str)+','+label_data['brixia3'].astype(str)+','+label_data['brixia4'].astype(str)+','+label_data['brixia5'].astype(str)+','+label_data['brixia6'].astype(str)
```

```python
label_data['num_split'] = 6
```

```python
label_data['file_id'] = label_data['Filename'].apply(lambda x: x.split('.')[0])
```

```python
label_data['subject_id'] = label_data['Subject']
```

```python
label_data['study_id'] = label_data['file_id']
```

```python
label_data['consensus'] = 0
```

```python
label_data=label_data[['subject_id','study_id','num_split','severity','file_id','consensus']]
```

```python
label_data
```

```python
brixia_metadata[brixia_metadata['Filename'].isin(brixia_consensus_data['Filename'])]
```

```python
# Find rows in brixia_metadata with matching Filename in brixia_consensus_data
partial_metadata = brixia_metadata[brixia_metadata['Filename'].isin(brixia_consensus_data['Filename'])]

# Merge partial_metadata with brixia_consensus_data on Filename
brixia_consensus_data = pd.merge(
    brixia_consensus_data, 
    partial_metadata[['Filename', 'Subject']],  # Only Filename and Subject columns from partial_metadata
    on='Filename', 
    how='left'  # Ensure all rows in brixia_consensus_data are kept
)

# Check the updated brixia_consensus_data
brixia_consensus_data
```

```python
# ModeA, ModeB, ModeC, ModeD, ModeE, ModeF -> str로 바꾸고 -> ,로 합치기
brixia_consensus_data['severity']=brixia_consensus_data['ModeA'].astype(str)+','+brixia_consensus_data['ModeB'].astype(str)+','+brixia_consensus_data['ModeC'].astype(str)+','+brixia_consensus_data['ModeD'].astype(str)+','+brixia_consensus_data['ModeE'].astype(str)+','+brixia_consensus_data['ModeF'].astype(str)
```

```python
brixia_consensus_data['num_split'] = 6
brixia_consensus_data['file_id'] = brixia_consensus_data['Filename'].apply(lambda x: x.split('.')[0])
brixia_consensus_data['subject_id'] = brixia_consensus_data['Subject']
brixia_consensus_data['study_id'] = brixia_consensus_data['file_id']
brixia_consensus_data['consensus'] = 1
brixia_consensus_data=brixia_consensus_data[['subject_id','study_id','num_split','severity','file_id','consensus']]
```

```python
brixia_consensus_data
```

```python
# label_data,brixia_consensus_data 합치기.
label_data = pd.concat([label_data,brixia_consensus_data])
```

```python
label_data.to_csv('/hdd/project/cylce_het/dataset/result_df/brixa_severity.csv',index=False)
```

# RALO

```python
# read, drop first row
ralo_data_df = pd.read_csv("/hdd/project/cylce_het/dataset/processed_df/ralo-dataset-metadata.csv", skiprows = 1)
```

```python
ralo_data_df=ralo_data_df[['Subject_ID','Right OPACITY', 'Left OPACITY']]
```

```python
# file_id는 인덱스
ralo_data_df['file_id'] =  ralo_data_df.index.astype(str)
```

```python
ralo_data_df['study_id'] = ralo_data_df['file_id']
```

```python
ralo_data_df['num_split'] = 2
```

```python
# NA가 포함된 행을 불러와서 시각화
ralo_data_df[ralo_data_df.isnull().any(axis=1)]
```

```python
# NA인 행 제거
ralo_data_df = ralo_data_df.dropna()
```

```python
ralo_data_df
```

```python
# Right OPACITY, Left OPACITY int로 바꾸고, str로 바꾸기. 그리고 둘을 이어붙이기.
ralo_data_df['Right OPACITY'] = ralo_data_df['Right OPACITY'].astype(int)
ralo_data_df['Left OPACITY'] = ralo_data_df['Left OPACITY'].astype(int)
# 둘을 , 로 이어붙이기
ralo_data_df['severity'] = ralo_data_df['Right OPACITY'].astype(str) + ',' + ralo_data_df['Left OPACITY'].astype(str)
```

```python
# Right OPACITY, Left OPACITY 컬럼 제거.
ralo_data_df = ralo_data_df.drop(['Right OPACITY', 'Left OPACITY'], axis = 1)
# Subject_ID를 subject_id로 바꾸기
ralo_data_df.rename(columns = {'Subject_ID':'subject_id'}, inplace = True)

ralo_data_df['consensus'] = 0
# 순서 바꾸기
ralo_data_df = ralo_data_df[['subject_id','study_id','num_split','severity','file_id','consensus']]
```

```python
ralo_data_df.to_csv("/hdd/project/cylce_het/dataset/result_df/ralo_severity.csv", index = False)
```
